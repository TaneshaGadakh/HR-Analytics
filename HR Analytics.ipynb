```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
hr1=pd.read_csv('HR_1.csv', encoding='unicode_escape')
```

```python
hr2=pd.read_csv('HR_2.csv', encoding='unicode_escape')
```

```python
hr2.rename(columns={'Employee ID':'EmployeeNumber'}, inplace=True)
```

```python
df=pd.merge(hr1, hr2, on='EmployeeNumber')
```

# Attrition Rate

```python
df['Attrition_Rate']=df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)
```

# Total Employees

```python
Total_Emp=df['EmployeeCount'].sum()
print(f'Total Employees = {Total_Emp/1000:.0f}K')
```

# Employee Attrition

```python
Emp_Att=df['Attrition_Rate'].sum()
print(f'Employee Attrition = {Emp_Att/1000:.2f}K')
```

# Attrition Rate

```python
Att_Rate=df['Attrition_Rate'].mean()*100
print(f'Attrition Rate = {Att_Rate}%')
```

# Gender Count

```python
ax=sns.countplot(x='Gender', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

# 1. Average Attrition rate for all Departments

```python
df['Department']=df['Department'].replace({'Research & Development':'R & D', 'Human Resources':'HR'})

dept_avg_att=df.groupby('Department')['Attrition_Rate'].mean()*100

plt.figure(figsize=(10,6))
ax=sns.barplot(x=dept_avg_att.index, y=dept_avg_att.values)
plt.title('Avg. Attrition Rate by Departments')
#plt.ylabel('Attrition Rate')
#plt.xlabel('Department')
#plt.xticks(rotation=45)

ax.bar_label(ax.containers[0], fmt='%.2f')

plt.tight_layout()
plt.show
```

# 2. Average Hourly Rate of Male Research Scientist

```python
avg_hourly_rate=df[(df['Gender']=='Male') & (df['JobRole']=='Research Scientist')]['HourlyRate'].mean()
print(f'Avg. Hourly Rate of Male Research Scientist = {avg_hourly_rate:.2f}')
```

#  3. Attrition Rate vs Monthly Income

```python
#Monthly Income Range

bins=[0, 15000, 30000, 45000, float('inf')]
lables=['1K - 15K', '16K - 30K', '31K - 45K', '46K+']
df['IncomeRange']=pd.cut(df['MonthlyIncome'], bins=bins, labels=lables, right=True)
```

```python
att_vs_income=df.groupby('IncomeRange')['Attrition_Rate'].mean()*100

plt.figure(figsize=(10,5))
ax=sns.barplot(x=att_vs_income.index, y=att_vs_income.values)
plt.title('Attrition Rate vs Monthly Income')
plt.xlabel('Monthly Income')

ax.bar_label(ax.containers[0], fmt='%.2f')

plt.tight_layout()
plt.show
```

# 4. Average Working Years for each Department

```python
avg_work_years=df.groupby('Department')['TotalWorkingYears'].mean()

plt.figure(figsize=(10,6))
ax=sns.barplot(x=avg_work_years.index, y=avg_work_years.values)
plt.title='Average Working Years by Department'

ax.bar_label(ax.containers[0], fmt='%.1f')

plt.tight_layout()
plt.show
```

# 5. Job Role vs Work Life Balance

```python
dt=df.pivot_table(index='JobRole', columns='WorkLifeBalance', aggfunc='size', fill_value=0)

plt.figure(figsize=(12,8))
ax=sns.heatmap(dt, annot=True, fmt="d", cmap="Greens", linewidths=0.5, cbar_kws={'label':'Count'})

#plt.title('Job Role vs Work Life Balance')
#plt.xlabel('Work Life Balance Ratings')
#plt.ylabel('Job Role')

plt.xticks(ticks=[0.5, 1.5, 2.5, 3.5], labels=['1', '2', '3', '4'], rotation=0)
plt.yticks(rotation=0)
plt.show()
```

# 6. Attrition Rate vs Year since Last Promotion Relation

```python
#Age Groups

bins=[0, 10, 20, 30, float('inf')]
lables=['1 - 10 Years', '11 - 20 Years', '21 - 30 Years', '31+ Years']
df['AgeGroup']=pd.cut(df['YearsSinceLastPromotion'], bins=bins, labels=lables, right=True)
```

```python
att_vs_ylp=df.groupby('AgeGroup')['Attrition_Rate'].mean()*100

plt.figure(figsize=(10,5))
ax=sns.barplot(x=att_vs_ylp.index, y=att_vs_ylp.values)
#plt.title('Attrition Rate vs Year since Last Promotion')

ax.bar_label(ax.containers[0], fmt='%.2f')

plt.tight_layout()
plt.show
```
